```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sn
```

```python
df = pd.read_csv(r"E:\downloads\archive (2)\Breast_Cancer.csv")
```

```python
df
```

```python
df = df.dropna()
```

```python
df
```

```python
sn.countplot(x = 'Marital Status' , data = df , edgecolor = 'black')
plt.title('Breast cancer')
plt.show
```

```python
sn.barplot(x = 'Age' , y = 'Grade', data = df , ci = False, edgecolor = 'black' )
plt.title('Breast cancer')
plt.ylabel('Grade')
plt.xlim(0,80)
plt.show
```

```python
sn.countplot(x = 'Progesterone Status' , data = df)
plt.title('Breast cancer')
plt.show
```

```python
plt.hist(df['Age'] , edgecolor = 'black')
plt.title('Breast Cancer')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show
```

```python
plt.scatter(df['Age'] , df['Tumor Size'], s = 2)
plt.title('Breast Cancer')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show
```

```python
sn.barplot(x = 'Status' , y = 'Tumor Size' , data = df , ci = False , edgecolor = 'black')
plt.ylim(0,40)
plt.title('Status and tumor size')
plt.show
```

```python
plt.scatter(x = df['Tumor Size'] , y = df['Survival Months'] , s = 7)
plt.xlabel('Tumor size')
plt.ylabel('Survival months')
plt.ylim(0,100)
plt.show
```
